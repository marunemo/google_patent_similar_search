```python
import requests, re
from bs4 import BeautifulSoup

'''similar search to google patent'''
def search_prior_art(prior_art: str) -> BeautifulSoup:
	URL = "https://patents.google.com"

	# recommend similar patent in google patent
	response = requests.get(f"{URL}/xhr/parse?text={prior_art}")
	most_similar = response.json()["results"][0]["result"]["id"]
	print("DEBUG prior art:", re.sub(r"patent\/(\w+)\/.*", r"\1", most_similar))

	# parse HTTP response by bs4
	response = requests.get(f"{URL}/{most_similar}")
	content = BeautifulSoup(response.content, "html.parser")

	return content

'''get description of citation in google patent'''
def get_specification(html: BeautifulSoup) -> str:
	description_section = html.select_one(".description")

	try:
		description = "".join([desc.text for desc in description_section.select("heading")[-1].next_siblings])
	except IndexError:
		description = description_section.text

	return description

'''get claim of citation in google patent'''
def get_claim(html: BeautifulSoup) -> list:
	claim_section = html.select_one(".claims")

	claim = claim_section.select("div[num]")
	claim = list(map(lambda tag: tag.text.replace("\n", " ").strip(), claim))

	return claim
```

```python
# example
pa = "US 20170119116"

html = search_prior_art(pa)
print(get_specification(html))
print(get_claim(html))
```
